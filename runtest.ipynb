寫出想訓練的 ./data/corpus2017.txt 及 ./data/corpus2018.txt

```
import data.data as dd

d2017 = dd.load_2017(5,5)
d2018 = dd.load_2018(1,5)
d2018_test = dd.load_2018_test()


```

```
def write_training_pairs(file, data):
  print("writing '%s' with %d pairs...\n" % (file, len(data['question'])))
  with open(file, 'wt') as out_file:
    for i in range(len(data['question'])):
      out_file.write(data['question'][i].strip() + '\r\n')
      out_file.write(data['answer'][i].strip() + '\r\n')


dataPath = '/content/drive/My Drive/Colab Notebooks/stc3/data'

filePath = dataPath + '/corpus2017.txt'
write_training_pairs(filePath, d2017)
    
filePath = dataPath + '/corpus2018.txt'
write_training_pairs(filePath, d2018)
```
